```python
import numpy as np
input_layer_neuron = 2
hidden_layer_neuron=3
output_neuron = 1

print("Input layer neurons = ",input_layer_neuron)
print("Hidden layer neurons = ",hidden_layer_neuron)
print("Output neurons = ",output_neuron)
```

```python
wh = np.random.uniform(size=(input_layer_neuron, hidden_layer_neuron))
print("Weights from input layer to hidden layer = ",wh)
```

```python
bh = np.random.uniform(size=(1,hidden_layer_neuron))
print("Bias between input layer and hidden layer = ", bh)
```

```python
wout = np.random.uniform(size=(hidden_layer_neuron, output_neuron))
print("Weights from hidden layer to output layer = ",wout)
```

```python
bout = np.random.uniform(size=(1,output_neuron))
print("Bias between hidden layer and output layer = ",bout)
```

```python
epoch = 7000
lr=0.1
x=np.array(([2,9],[1,5],[3,6]),dtype=float)
y=np.array(([92],[86],[89]),dtype=float)
x=x/np.amax(x,axis=0)
print(x)
```

```python
def sigmoid(x):
  return 1/(1+np.exp(-x))

def derivatives_sigmoid(x):
  return x*(1-x)

for i in range (epoch):
  hinp1 = np.dot(x,wh)
  hinp=hinp1+bh
  h1layer_act = sigmoid(hinp)
  outinp1 = np.dot(h1layer_act, wout)
  outinp = outinp1+bout
  output=sigmoid(outinp)
  eo = y-output
  outgrad = derivatives_sigmoid(output)
  d_output = eo * outgrad
  eh = d_output.dot(wout.T)
  hiddengrad = derivatives_sigmoid(h1layer_act)
  d_hiddenlayer = eh * hiddengrad
  wout+=h1layer_act.T.dot(d_output)*lr
  wht = x.T.dot(d_hiddenlayer)*lr

  print("Input: \n",str(x))
  print("Actual output: \n",str(y))
  print("Predicted output: ",output)
```
